## PREPROCESSNG 

```python
## COPY OF dATA Set


email_df = df.copy()

# Display first five rows
email_df.head()
```

```python
print("Dataset Shape:", email_df.shape)
```

```python

print("Duplicate Emails:", email_df.duplicated().sum())

email_df.drop_duplicates(inplace=True)

print("Shape after removing duplicates:", email_df.shape)
```

```python
##  Mssing VAlues
email_df.isnull().sum()
```

###  Extract Subject 

```python

email_df["Subject"] = email_df["message"].str.extract(r"Subject:(.*)")
```

## Extract Sender

```python
email_df["From"] = email_df["message"].str.extract(r"From:(.*)")
```

### Extract Date

```python
email_df["Date"] = email_df["message"].str.extract(r"Date:(.*)")
```

### Extract Recipient


```python

email_df["To"] = email_df["message"].str.extract(r"To:(.*)")
```

### View Extracted Columns

```python
email_df[["From", "To", "Subject", "Date"]].head()
```

### : Extract the Email Body

```python
email_df["Body"] = email_df["message"].str.split("X-FileName:").str[-1]

email_df["Body"].head()
```

### Remove Leading and Trailing Spaces

```python

email_df["Body"] = email_df["Body"].str.strip()

email_df["Body"].head
```

### Convert Text to Lowercase

```python
email_df["Body"] = email_df["Body"].str.lower()

email_df["Body"].head()
```

### Remove Email Addresses

```python
email_df["Body"] = email_df["Body"].str.replace(
    r"\S+@\S+", "", regex=True
)

email_df["Body"].head()
```

### Remove URLs

```python

email_df["Body"] = email_df["Body"].str.replace(
    r"http\S+|www\S+", "", regex=True
)

email_df["Body"]
```

### 

## Remove Numbers 

```python
email_df["Body"] = email_df["Body"].str.replace(
    r"\d+", "", regex=True
)

email_df["Body"].head()
```

### Remove Punctuation

```python
email_df["Body"] = email_df["Body"].str.replace(
    r"[^\w\s]", "", regex=True
)

email_df["Body"].head()
```

### : Remove Extra Spaces

```python
email_df["Body"] = email_df["Body"].str.replace(
    r"\s+", " ", regex=True
).str.strip()

email_df["Body"].head()
```

```python
print("Original Email:\n")
print(df["message"][0])

print("\n" + "="*100 + "\n")

print("Cleaned Email:\n")
print(email_df["Body"][0])
```

### Remove Empty Email Bodies

```python
email_df = email_df[email_df["Body"].str.strip() != ""]

# Reset index
email_df.reset_index(drop=True, inplace=True)

print("Dataset Shape:", email_df.shape)
```

### Calculate Email Length

```python

email_df["Character_Count"] = email_df["Body"].apply(len)

email_df[["Body","Character_Count"]].head()
```

### Calculate Word Count

```python
email_df["Word_Count"] = email_df["Body"].apply(lambda x: len(x.split()))

email_df[["Word_Count"]].head()
```

### : Visualize Word Count Distribution

```python
plt.figure(figsize=(10,5))

plt.hist(email_df["Word_Count"], bins=40)

plt.title("Distribution of Email Word Count")
plt.xlabel("Number of Words")
plt.ylabel("Frequency")
```

### Remove Very Short Emails

```python
email_df = email_df[email_df["Word_Count"] >= 10]

email_df.reset_index(drop=True, inplace=True)

print("Dataset Shape:", email_df.shape)
```

### Save Cleaned Dataset

```python
email_df.to_csv("cleaned_emails.csv", index=False)

print("Cleaned dataset saved successfully.")
```

```python
from transformers import BartTokenizer, BartForConditionalGeneration
import torch
```

```python
model_name = "sshleifer/distilbart-cnn-12-6"

tokenizer = BartTokenizer.from_pretrained(model_name)
model = BartForConditionalGeneration.from_pretrained(model_name)

print("BART model loaded successfully.")
```

### Select an Email

```python
email = email_df["Body"][0]

print(email[:1000])  
```

### Tokenize the Email

```python

inputs = tokenizer(
    email,
    max_length=1024,
    truncation=True,
    return_tensors="pt"
)
```

### Genrate Summary 

```python
summary_ids = model.generate(
    inputs["input_ids"],
    max_length=120,
    min_length=30,
    num_beams=4,
    early_stopping=True
)
```

### Decode the Summary

```python
summary = tokenizer.decode(
    summary_ids[0],
    skip_special_tokens=True
)

print("Generated Summary:\n")
print(summary)
```

```python
print("=" * 100)
print("ORIGINAL EMAIL")
print("=" * 100)

print(email)

print("\n")

print("=" * 100)
print("GENERATED SUMMARY")
print("=" * 100)

print(summary)
```

```python
# Save the generated summary in the DataFrame

email_df.loc[0, "Generated_Summary"] = summary

# Display the result
email_df[["Body", "Generated_Summary"]].head()
```
